```python
import pandas as pd

# Load your CSV
df = pd.read_csv("results_variants_effcientnetv2s.csv")

# ---- TABLE 1: Accuracy summary ----
accuracy_table = df.pivot_table(
    index="variant",
    columns="model",
    values="accuracy"
).round(3)

# ---- TABLE 2: Speed summary (choose one metric) ----
# Option A: time per sample
speed_table_sample = df.pivot_table(
    index="variant",
    columns="model",
    values="time_per_sample_sec"
).round(6)

# Option B: time per batch
speed_table_batch = df.pivot_table(
    index="variant",
    columns="model",
    values="time_per_batch_sec"
).round(4)

# Print results
print("\n=== Accuracy Table ===")
print(accuracy_table)

print("\n=== Time per Sample (sec) ===")
print(speed_table_sample)

print("\n=== Time per Batch (sec) ===")
print(speed_table_batch)
```
