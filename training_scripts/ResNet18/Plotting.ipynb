```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("results/ResNet18_metrics.csv")



# Get unique methods
methods = df["method"].unique()

for method in methods:
    df_m = df[df["method"] == method].sort_values("epoch")

    epochs = df_m["epoch"]

    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    #for loss
    axes[0].plot(epochs, df_m["train_loss"], label="Train Loss")
    axes[0].plot(epochs, df_m["val_loss"], label="Val Loss")
    axes[0].set_title(f"{method}: Loss")
    axes[0].set_xlabel("Epoch")
    axes[0].set_ylabel("Loss")
    axes[0].legend()

    # for accuracy
    axes[1].plot(epochs, df_m["train_acc"], label="Train Acc")
    axes[1].plot(epochs, df_m["val_acc"], label="Val Acc")
    axes[1].set_title(f"{method}: Accuracy")
    axes[1].set_xlabel("Epoch")
    axes[1].set_ylabel("Accuracy")
    axes[1].legend()

    plt.tight_layout()
    plt.savefig(f"{method}_metrics.png", dpi=300)
    plt.show()
```
